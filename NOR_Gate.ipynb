```python
import numpy as np

# Create a Perceptron class
class Perceptron:
    def __init__(self, weights, bias):
        self.weights = np.array(weights)
        self.bias = bias

    # Prediction function
    def predict(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return 1 if summation >= 0 else 0

# Input combinations
inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# NOR Gate
nor_gate = Perceptron(weights=[-1, -1], bias=0.5)

print("NOR Gate")

for x in inputs:
    print(f"{x} -> {nor_gate.predict(x)}")
```
