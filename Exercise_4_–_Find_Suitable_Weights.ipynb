```python
import numpy as np

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = np.array(weights)
        self.bias = bias

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return 1 if summation >= 0 else 0

inputs = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

bias = -0.5

# AND Gate
and_gate = Perceptron(weights=[0.5,0.5], bias=bias)

# OR Gate
or_gate = Perceptron(weights=[1,1], bias=bias)

print("AND Gate")
for x in inputs:
    print(f"{x} -> {and_gate.predict(x)}")

print("\nOR Gate")
for x in inputs:
    print(f"{x} -> {or_gate.predict(x)}")
```
