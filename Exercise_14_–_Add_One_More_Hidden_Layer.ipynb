```python
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# XOR Dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,1,1,0])

# -----------------------------
# Model 1: One Hidden Layer
# -----------------------------
model1 = MLPClassifier(
    hidden_layer_sizes=(2,),   # One hidden layer with 2 neurons
    activation='logistic',
    solver='lbfgs',
    max_iter=1000,
    random_state=42
)

model1.fit(X, y)

pred1 = model1.predict(X)

acc1 = accuracy_score(y, pred1)

# -----------------------------
# Model 2: Two Hidden Layers
# -----------------------------
model2 = MLPClassifier(
    hidden_layer_sizes=(2,2),   # Two hidden layers (2 neurons each)
    activation='logistic',
    solver='lbfgs',
    max_iter=1000,
    random_state=42
)

model2.fit(X, y)

pred2 = model2.predict(X)

acc2 = accuracy_score(y, pred2)

# -----------------------------
# Results
# -----------------------------
print("===== One Hidden Layer =====")
print("Predicted Output :", pred1)
print("Accuracy         :", acc1 * 100, "%")
print("Loss             :", model1.loss_)

print("\n===== Two Hidden Layers =====")
print("Predicted Output :", pred2)
print("Accuracy         :", acc2 * 100, "%")
print("Loss             :", model2.loss_)
```
