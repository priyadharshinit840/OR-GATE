```python
import numpy as np

# Perceptron Class
class Perceptron:
    def __init__(self, weights, bias):
        self.weights = np.array(weights)
        self.bias = bias

    # Sigmoid Activation
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    # ReLU Activation
    def relu(self, x):
        return max(0, x)

    # Prediction using Sigmoid
    def predict_sigmoid(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return self.sigmoid(summation)

    # Prediction using ReLU
    def predict_relu(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return self.relu(summation)

# Inputs
inputs = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# OR Gate Parameters
perceptron = Perceptron(weights=[1,1], bias=-0.5)

print("Input\tSigmoid\t\tReLU")

for x in inputs:
    s = perceptron.predict_sigmoid(x)
    r = perceptron.predict_relu(x)
    print(f"{x}\t{s:.4f}\t\t{r:.2f}")
```
