```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# XOR Dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,1,1,0])

# Epoch values
epochs = [10, 30, 50, 100]

accuracy = []
loss = []

for epoch in epochs:

    model = MLPClassifier(
        hidden_layer_sizes=(2,),      # One hidden layer with 2 neurons
        activation='logistic',
        solver='lbfgs',
        max_iter=epoch,
        random_state=42
    )

    model.fit(X, y)

    pred = model.predict(X)

    accuracy.append(accuracy_score(y, pred) * 100)
    loss.append(model.loss_)

# Display Results
results = pd.DataFrame({
    "Epochs": epochs,
    "Accuracy (%)": accuracy,
    "Loss": loss
})

print(results)

# Accuracy Plot
plt.figure(figsize=(6,4))
plt.plot(epochs, accuracy, marker='o')
plt.title("Epochs vs Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.grid(True)
plt.show()

# Loss Plot
plt.figure(figsize=(6,4))
plt.plot(epochs, loss, marker='o')
plt.title("Epochs vs Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```
